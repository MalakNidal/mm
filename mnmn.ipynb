```python
##Distribution of Market Value
sns.set()
_=plt.hist(df_medterm['market_value'])
_=plt.xlabel('market value')
_=plt.ylabel('frequence')
plt.show()
```

```python
##Market Value with Age
_= sns.swarmplot(x='age',y='market_value', data=df_medterm)
_=plt.xlabel('age')
_=plt.ylabel('market_value')
plt.show()
```

```python
##The market value is big when th player ages between 22 to 31
```
